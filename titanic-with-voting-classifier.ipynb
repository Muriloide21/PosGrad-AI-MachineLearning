```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.head()
```

```python
X = train_data.drop('Survived', axis=1)
X.columns
```

```python
# from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

# model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
# model.fit(X, y)
# predictions = model.predict(X_test)

# output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
# output.to_csv('submission.csv', index=False)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier, AdaBoostClassifier, VotingClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

from sklearn.model_selection import GridSearchCV

# # Defina os hiperparâmetros a serem testados
# param_grid = {
#     'C': [0.001, 0.01, 0.1, 1, 10],
#     'kernel': ['linear', 'rbf', 'poly'],
#     'gamma': ['scale', 'auto']
# }

# # Crie o RandomForestClassifier
# svc_classifier = SVC(random_state=1)

# # Use GridSearchCV para encontrar os melhores hiperparâmetros
# grid_search = GridSearchCV(svc_classifier, param_grid, cv=5, scoring='accuracy')
# grid_search.fit(X, y)

# # Obtenha os melhores hiperparâmetros
# best_params = grid_search.best_params_
# print("Melhores Hiperparâmetros SVC:", best_params)


# Depois de rodar o Grid Search para cada um dos classificadores e ajustar os hiperparâmetros
rf_classifier = RandomForestClassifier(n_estimators=150, max_depth=5, random_state=1, min_samples_leaf=2, min_samples_split=5)
gb_classifier = GradientBoostingClassifier(n_estimators=150, learning_rate=0.05, max_depth=3, min_samples_leaf=4, min_samples_split=2, random_state=1)
et_classifier = ExtraTreesClassifier(n_estimators=50, max_depth=5, min_samples_leaf=2, min_samples_split=2, random_state=1)
ada_classifier = AdaBoostClassifier(n_estimators=150, learning_rate=0.1, random_state=1)
lr_classifier = LogisticRegression(C=100,random_state=1)
svc_classifier = SVC(C=1, gamma='scale', kernel='rbf', probability=True, random_state=1)

# Voting Classifier
voting_classifier = VotingClassifier(estimators=[
    ('rf', rf_classifier),
    ('gb', gb_classifier),
    ('et', et_classifier),
    ('ada', ada_classifier),
    ('lr', lr_classifier),
    ('svc', svc_classifier)
], voting='hard')

scores_voting = cross_val_score(voting_classifier, X, y, cv=5)
print("Voting Classifier Accuracy: %0.2f (+/- %0.2f)" % (scores_voting.mean(), scores_voting.std() * 2))

# Treinando o modelo final para o Voting Classifier
voting_classifier.fit(X, y)

# Predições
predictions_voting = voting_classifier.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions_voting})
output.to_csv('submission.csv', index=False)
```
